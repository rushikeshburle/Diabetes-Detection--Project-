```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix

def diabetesfeatures(Pregnancies, Glucose, Blood_Pressue, Skin_thickness, Insulin, BMI, Dibetes_Predigree_Function, Age):
    dataset = pd.read_csv(r"C:\Users\karis\OneDrive\Desktop\Diabetes.csv.csv")

    x=dataset.drop('Outcome',axis=1)
    y=dataset['Outcome']

    classifier=KNeighborsClassifier()
	
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
    
    classifier.fit(x_train,y_train)

    predictedresult= classifier.predict(x_test)
  
    print("Result expected ", y_test)
    print("Result predicted by ML",predictedresult)
    
    Accuracy=metrics.accuracy_score(y_test, predictedresult)
    
    print((Accuracy)*100)
    
 
    
    
    Prediction=classifier.predict([[Pregnancies, Glucose, Blood_Pressue, Skin_thickness, Insulin, BMI, Dibetes_Predigree_Function, Age]])
    if Prediction==0:
        print("No Chances of Diabetes")
    elif Prediction==1:
        print("Chances of Diabetes")
   
    
def main():

    print("Enter the Pregnancies")
    Pregnancies=float(input())
    print("Enter the Glucose")
    Glucose=float(input())
    print("Enter the Blood_Pressue")
    Blood_Pressue=float(input())
    print("Enter the Skin_thickness")
    Skin_thickness=float(input())
    print("Enter the Insulin")
    Insulin=float(input())
    print("Enter the BMI")
    BMI=float(input())
    print("Enter the Dibetes_Predigree_Function")
    Dibetes_Predigree_Function=float(input())
    print("Enter the Age")
    Age=float(input())
    diabetesfeatures(Pregnancies,Glucose,Blood_Pressue,Skin_thickness,Insulin,BMI,Dibetes_Predigree_Function,Age)

if __name__=="__main__":
	main()

```
